## Data Cleaning 

### Fill Missing Values 

```python
ec['type'] = ec['type'].fillna(ec['type'].mode()[0])
```

```python
ec[ec["Turnout"].isnull()]
```

```python
ec[ec["year"].isnull()]
```

```python
ec = ec.dropna(axis=0, subset=['Turnout', 'margin%', 'year'])
```

```python
ec.isnull().sum()
```

### Remove unwanted Signs 

```python
ec['Turnout'] = ec['Turnout'].str.replace('%', '')
ec['margin%'] = ec['margin%'].str.replace('%', '')
```

```python
ec.head()
```

```python
ec[ec["Turnout"] == '-']
```

```python
ec = ec.drop(ec[ec['Turnout'] == '-'].index)
```

```python
ec.shape
```

```python
ec['electors'] = ec.electors.apply(lambda x: x.strip(','))

ec['electors'] = ec.electors.apply(lambda x: x.replace(',',''))
```

```python
ec['votes'] = ec.votes.apply(lambda x: x.strip(','))

ec['votes'] = ec.votes.apply(lambda x: x.replace(',',''))
```

```python
ec['margin'] = ec.margin.apply(lambda x: x.strip(','))

ec['margin'] = ec.margin.apply(lambda x: x.replace(',',''))
```

```python
ec.head()
```

### Change Datatypes

```python
ec['no'] = ec['no'].astype(int)
ec['electors'] = ec['electors'].astype(int)
ec['votes'] = ec['votes'].astype(int)
ec['Turnout'] = ec['Turnout'].astype(float)
ec['margin'] = ec['margin'].astype(int)
ec['margin%'] = ec['margin%'].astype(float)
ec['year'] = ec['year'].astype(int)
```

```python
ec.rename(index=str, columns={'Turnout': 'turnout%'}, inplace=True)
```

## Haryana Elections

```python
Hyr= (1967, 1971, 1977, 1980, 1984, 1989, 1991, 1996, 1998, 1999,
            2004, 2009, 2014, 2019)
```

```python
Hr19 = styr.get_group(('Haryana',2019))
Hr19
```

```python
sns.set(rc={'figure.figsize':(14,7)})
Hr19.plot(x='Pc_name',y=['margin','votes','electors']);
```

```python
# creating a coustom colour pallete

top='#9E1C00'
mid='#F8E506'
low='#1CA4A4'
```

```python
col=[low,mid,top]
```

```python
Hr19.plot(x='Pc_name',y=['margin','electors','votes'],kind='bar',figsize=(16,8.5),width=.9, color = col)
plt.show;
```

```python
Hr19.plot(x='Pc_name',y=['turnout%'],kind='bar',figsize=(10,6),width=.9, color = col)
plt.show;
```

```python
P19=Hr19.groupby('party').count()
```

```python
#combining the above visualisations to show the results in kerala over the years
print('Loksabha elections in Haryana - an overview')
for yr in Hyr:
    n1=pd.DataFrame()
    n1=styr.get_group(('Haryana',yr))
    
#plotting bar chart
    n1.plot(x='Pc_name',y=['margin','electors','votes'],kind='bar',figsize=(16,8.5),width=.9, color=col)
    plt.show
    print(yr)
    
    P19=n1.groupby('party').count()
    
    explode=[]
    exp=P19.index
    for i in exp:
        explode.append(0.1)
    explode.pop()
    explode.append(0)
 
    # Creating autocpt arguments
    def func(pct, allvalues):
        absolute = int(pct / 100.*np.sum(allvalues))
        return "{:d} ".format( absolute)

    # Creating plot
    fig, ax = plt.subplots(figsize =(20, 7))
    wedges, texts, autotexts = ax.pie(P19.votes, 
                                  autopct = lambda pct: func(pct, P19.votes),
                                  explode = explode, 
                                  labels = P19.index,
                                  shadow = True,
                                  startangle = 180,
                                  textprops = dict(color ="#173A40"))
    
     # Adding legend
    ax.legend(wedges, exp,
          title ="Parties",
          loc ="center left",
          bbox_to_anchor =(1, 0, 0.5, 1))
 

    plt.setp(autotexts, size = 8, weight ="bold")
    ax.set_title("Political parties and their seats in "+str(yr))
  
    # show plot
    plt.show()
```

## Performance of National Parties over years

```python
seatyr=ec.groupby(['party','year']).count()
```

```python
md = seatyr.reset_index()
md.head()
```

```python
seats=md.drop(['Pc_name','type','candidate_name','electors','turnout%','margin','margin%','state','votes'],axis=1)
seats.head()
```

```python
seats.rename(columns={'no':'seats'},inplace=True)
seats
```

```python
national_parties=['Indian National Congress','Bharatiya Janta Party','All India Trinamool Congress','Bahujan Samaj Party',
                  'Communist Party Of India','Communist Party Of India (MARXIST)','Nationalist Congress Party']
```

```python
Nat_P=seats.loc[seats['party'].isin(national_parties)]
```

```python
tmc='#FFF900'
bsp='#31a5fb'
bjp='#ff8c00'
cpi='#ff009f'
cpm='#ff0000'
inc='#083df3'
ncp='#4e2c81'

lstcol=[tmc,bsp,bjp,cpi,cpm,inc,ncp]
```

```python
fig, ax = plt.subplots(figsize=(12, 7))
sns.lineplot(data=Nat_P, x="year", y="seats", hue="party", palette=lstcol, linewidth = 2.5)
plt.legend( bbox_to_anchor=(1.02, 1),loc='upper left', borderaxespad=0, title='Party')
plt.title('Trend of Loksabha seats over the years of Recognized National Parties')
plt.show;
```
